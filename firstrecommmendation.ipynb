```python
import pandas as pd

data = pd.read_csv("channel dataset/data.csv")
print( data.columns )
print( data.head() )
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```python
featurematrix = cv.fit_transform(data["Genre"].to_numpy())
print( cv.get_feature_names())
print( featurematrix.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity( featurematrix)
print( similarity_matrix)
```

```python
index_of_show_user_watched = data[ data.Show == "Avatar"]["Index"].values[0]

print( index_of_show_user_watched)
```

```python
contentsimilararray = similarity_matrix[index_of_show_user_watched ] 

list_of_pair = list( enumerate(contentsimilararray ) )

print( list_of_pair)
```

```python
def mysortfunc(item):
    return item[1]

list_of_pair.sort( reverse =True ,key = mysortfunc )
print( list_of_pair )
```

```python
for show_tuple in list_of_pair[0:5]:
    currow = data[ data.index == show_tuple[0] ] 
    print( currow["Show"].values[0] , " ", currow["Channel"].values[0]," ", currow["Time"].values[0])
```
